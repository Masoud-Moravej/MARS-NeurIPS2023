```python
import pickle
from numpy import *
from matplotlib.pyplot import *
```

```python
file_to_read = open("stored_simulation/fig2a.pkl", "rb")
dict_data_fig2a = pickle.load(file_to_read)

fig, ax = subplots(figsize=(8, 6))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
e_colors = ['#8db8d9', '#ffb27d', '#9ed89e', '#de8280']

legen_names = ['Vanila-UCB','Normal-TS','Bootstrapped UCB','MARS']


k = 0

dict_data = dict_data_fig2a

for policy in dict_data:
    tsav = dict_data[policy][0]
    varRegret = dict_data[policy][2]
    meanRegret = dict_data[policy][1]

    ax.plot(1+tsav, meanRegret, color = colors[k], label=legen_names[k])
    ax.fill_between(1+tsav, subtract(meanRegret, varRegret), add(meanRegret, varRegret), color=e_colors[k], alpha=0.5)
    
    xlabel('Round')
    ylabel('Regret')
    k = k+1

legend()
```
